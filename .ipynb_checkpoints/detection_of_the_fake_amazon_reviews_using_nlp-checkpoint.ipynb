```python
import pandas as pd

# Convert reviewTime to datetime
df['reviewTime'] = pd.to_datetime(df['reviewTime'], errors='coerce')

df['vote'] = df['vote'].fillna(0)

df['vote'] = df['vote'].astype(str).str.replace(',', '', regex=False)

df['vote'] = pd.to_numeric(df['vote'], errors='coerce').fillna(0).astype(int)

df['reviewerName'] = df['reviewerName'].fillna("Unknown")

df['reviewText'] = df['reviewText'].fillna("")

df['verified'] = df['verified'].astype(int)

# Ovaj dio čisti podatke i priprema kolone:
# 1. 'reviewTime' pretvara u datum i vrijeme.
# 2. 'vote' prazne vrijednosti postavlja na 0 i pretvara u cijele brojeve.
# 3. Prazna imena recenzenata ('reviewerName') zamjenjuje sa "Unknown".
# 4. Prazne recenzije ('reviewText') zamjenjuje sa praznim stringom.
# 5. 'verified' pretvara iz True/False u 1/0 radi lakše obrade.
```

```python
# Tektualni

df['review_length'] = df['reviewText'].str.len()
df['summary_length'] = df['summary'].str.len()
df['word_count'] = df['reviewText'].str.split().str.len()
```

```python
# review po useru
df['reviews_by_user'] = df.groupby('reviewerID')['reviewerID'].transform('count')

# Avg
df['avg_user_rating'] = df.groupby('reviewerID')['overall'].transform('mean')

# Standard deviation of user ratings (fake reviewers often rate everything 5)
df['std_user_rating'] = df.groupby('reviewerID')['overall'].transform('std').fillna(0)
```

```python
df['review_year'] = df['reviewTime'].dt.year
df['review_month'] = df['reviewTime'].dt.month

# Days between reviews per reviewer (bots often post many reviews quickly)
# dani izmedju reviews/reviewer - botovi precesto postavljaju reviews
df['unixReviewTime'] = df['unixReviewTime'].astype(int)
df['time_diff'] = df.groupby('reviewerID')['unixReviewTime'].diff().fillna(0)
```

```python
# Number of reviews on the same product
df['reviews_per_product'] = df.groupby('asin')['asin'].transform('count')
```

```python
df['label'] = df['verified'].apply(lambda x: 0 if x == 1 else 1)
```

```python
df['label'] = df['verified'].apply(lambda x: 0 if x == 1 else 1)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

ps = PorterStemmer()
stop = set(stopwords.words('english'))

def clean_text(text):
    text = re.sub(r"[^a-zA-Z ]", "", text)
    words = text.lower().split()
    words = [w for w in words if w not in stop]
    words = [ps.stem(w) for w in words]
    return " ".join(words)

df['cleaned_text'] = df['reviewText'].apply(clean_text)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(
    df['cleaned_text'],
    df['label'],
    test_size=0.2,
    random_state=42
)

tfidf = TfidfVectorizer(max_features=5000)
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec = tfidf.transform(X_test)

model = LogisticRegression(max_iter=200)
model.fit(X_train_vec, y_train)

pred = model.predict(X_test_vec)
print(classification_report(y_test, pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from wordcloud import WordCloud
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
import nltk
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

nltk.download('stopwords')
nltk.download('punkt')


# You can increase head(10) → head(100) or remove .head() for full dataset
df = pd.read_json(
    "/Users/mirsad/detection_of_fake_amazon_reviews_using_nlp/Software_5.json",
    lines=True
).head(100)   # use more rows for better LDA/topic modeling

df.columns()

df['reviewText'] = df['reviewText'].fillna("").astype(str)

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()                              # lowercase
    text = re.sub(r'\W+', ' ', text)                 # remove special characters
    tokens = text.split()                            # tokenization
    tokens = [w for w in tokens if w not in stop_words]  # remove stopwords
    return " ".join(tokens)

df['clean_text'] = df['reviewText'].apply(preprocess_text)

analyzer = SentimentIntensityAnalyzer()

df['sentiment_score'] = df['clean_text'].apply(
    lambda x: analyzer.polarity_scores(x)['compound']
)

def classify_sentiment(score):
    if score > 0.05:
        return "Positive"
    elif score < -0.05:
        return "Negative"
    else:
        return "Neutral"

df['sentiment'] = df['sentiment_score'].apply(classify_sentiment)


plt.figure(figsize=(8, 5))
sns.countplot(x='sentiment', data=df, order=['Negative', 'Neutral', 'Positive'])
plt.title("Sentiment Classification")
plt.xlabel("Sentiment Category")
plt.ylabel("Count")
plt.show()

# Sentiment Score Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['sentiment_score'], bins=30, kde=True)
plt.title("Sentiment Score Distribution")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['overall'], y=df['sentiment_score'], alpha=0.5)
plt.title("Ratings vs. Sentiment Score")
plt.xlabel("Star Rating (overall)")
plt.ylabel("Sentiment Score (VADER compound)")
plt.show()


def generate_wordcloud(sentiment_label):
    text = " ".join(df[df['sentiment'] == sentiment_label]['clean_text'])
    if len(text.strip()) == 0:
        print(f"No text found for sentiment = {sentiment_label}, skipping wordcloud.")
        return
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(8, 4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud - {sentiment_label}")
    plt.show()

generate_wordcloud('Positive')
generate_wordcloud('Neutral')
generate_wordcloud('Negative')

df['timestamp'] = pd.to_datetime(df['unixReviewTime'], unit='s')

df_time = df.set_index('timestamp').sort_index()

# Monthly average sentiment
plt.figure(figsize=(10, 5))
df_time['sentiment_score'].resample('M').mean().plot()
plt.title("Sentiment Trend Over Time (Monthly Avg)")
plt.ylabel("Average Sentiment Score")
plt.xlabel("Time")
plt.show()


def get_ngrams(text, n):
    tokens = word_tokenize(text)
    n_grams = list(ngrams(tokens, n))
    return [" ".join(gram) for gram in n_grams]

df['bigrams'] = df['clean_text'].apply(lambda x: get_ngrams(x, 2))
df['trigrams'] = df['clean_text'].apply(lambda x: get_ngrams(x, 3))


bigram_list = [bigram for sublist in df['bigrams'] for bigram in sublist]
bigram_freq = Counter(bigram_list).most_common(15)

bigram_df = pd.DataFrame(bigram_freq, columns=['Bigram', 'Count'])

plt.figure(figsize=(10, 5))
sns.barplot(y='Bigram', x='Count', data=bigram_df)
plt.title("Most Common Bigrams in Reviews")
plt.xlabel("Frequency")
plt.ylabel("Bigram")
plt.show()

# Most Common Trigrams
trigram_list = [trigram for sublist in df['trigrams'] for trigram in sublist]
trigram_freq = Counter(trigram_list).most_common(15)

trigram_df = pd.DataFrame(trigram_freq, columns=['Trigram', 'Count'])

plt.figure(figsize=(10, 5))
sns.barplot(y='Trigram', x='Count', data=trigram_df)
plt.title("Most Common Trigrams in Reviews")
plt.xlabel("Frequency")
plt.ylabel("Trigram")
plt.show()

# =========================
# LDA TOPIC MODELING
# =========================
# Work on clean_text column
vectorizer = CountVectorizer(max_features=5000, stop_words='english')
text_matrix = vectorizer.fit_transform(df['clean_text'])

lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(text_matrix)

words = vectorizer.get_feature_names_out()

print("\nTOPICS FOUND BY LDA:\n")
for topic_idx, topic in enumerate(lda.components_):
    top_words = [words[i] for i in topic.argsort()[-10:]]
    print(f"Topic {topic_idx + 1}: {', '.join(top_words)}")
```

import pandas as pd

df = pd.DataFrame({
    'verified': [True, True, False, True],
    'rating': [5, 5, 4, 5]
})

df.value_counts()

```python
# mismatch feature 

# Normalize rating to [-1, 1]
df['rating_norm'] = (df['overall'] - 3) / 2

# Absolute mismatch
df['rating_sentiment_gap'] = abs(df['rating_norm'] - df['sentiment_score'])

# Flag strong mismatch
df['rating_sentiment_mismatch'] = df['rating_sentiment_gap'] > 0.7
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='verified', y='rating_sentiment_gap', data=df)
plt.title("Rating–Sentiment Gap vs Verified Purchase")
plt.xlabel("Verified Purchase (0=No, 1=Yes)")
plt.ylabel("Gap")
plt.show()
```

```python
# Create flags
df['word_count'] = df['reviewText'].astype(str).str.split().str.len()

df['is_very_short'] = df['word_count'] < 20
df['is_very_long'] = df['word_count'] > 300
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['word_count'], bins=50)
plt.title("Distribution of Review Length (Word Count)")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.show()
```

```python
df['length_anomaly'] = (
    (df['word_count'] < 15) | (df['word_count'] > 300)
).astype(int)

sns.countplot(x='length_anomaly', data=df)
plt.title("Extreme Review Length Anomalies")
plt.show()
```

```python
# bot ponasanje

df['unixReviewTime'] = df['unixReviewTime'].astype(int)
df = df.sort_values(by=['reviewerID', 'unixReviewTime'])
df['time_diff'] = df.groupby('reviewerID')['unixReviewTime'].diff()
df['time_diff'] = df['time_diff'].fillna(0)
df['fast_reviewer'] = (df['time_diff'] < 60).astype(int)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='fast_reviewer', data=df)
plt.title("Fast Reviewer Behavior (Burst Posting)")
plt.xlabel("Posted within 1 minute of previous review")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='vote',
    y='word_count',
    hue='verified',
    data=df,
    alpha=0.6
)
plt.title("Helpfulness Votes vs Review Length")
plt.xlabel("Votes")
plt.ylabel("Word Count")
plt.show()
```

```python
def lexical_diversity(text):
    words = text.split()
    if len(words) == 0:
        return 0
    return len(set(words)) / len(words)

df['lexical_diversity'] = df['clean_text'].apply(lexical_diversity)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='verified', y='lexical_diversity', data=df)
plt.title("Lexical Diversity vs Verified Purchase")
plt.show()
```
